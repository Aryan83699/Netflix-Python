```python
# Importing the necessary libraries

import pandas as pd 
import numpy as np
```

```python
# Reading the csv files

df=pd.read_csv('netflix_titles.csv')
```

```python
# Shape of the raw dataset 

print(f" It has {df.shape[0]} no. of rows and {df.shape[1]} columns")
```

```python
# Overview of the Dataset

df.sample(7)
```

```python
df.info()
```

```python
# 1. Count the Number of TV Shows and Movies 

df.groupby('type')['show_id'].size()
```

```python
# 2. Finding the most common ratings for movies and tv Shows 

df_grouped = df.groupby(['type', 'rating']).size().reset_index(name='frequency')

# Step 2: Rank within each 'type' by descending frequency
df_grouped['rankings'] = df_grouped.groupby('type')['frequency'].rank(method='dense', ascending=False)

# Step 3: Filter only rank = 1 (most frequent rating per type)
result = df_grouped[df_grouped['rankings'] == 1][['type', 'rating']]

print(result)
```

```python
# 3. List all the movies released in an specific year (eg. 2020) 

df[(df.type=='Movie') & (df.release_year==2020)]
```

```python
# 4. Find the top 5 Countries with most content on Netflix 

# step .1 :- creating copy of the dataset
df1=df.copy()

# step 2 :- splitting the string into list of arrays
df1['country_list']=df['country'].str.split(',')

# step 3 :- unnesting the list and triming the unnecessary spaces
df2=df1.explode('country_list')['country_list'].str.strip()

# step 4 :- taking the count according to the categories
df2.value_counts().head(5)
```

```python
# 5. Identify the Longest running movie 

# step 1 :- removing the 'MIN' and converting the duration into Integer
max_duration=df[df.type=='Movie']['duration'].str.replace('min','').astype('Int64').max()

# step 2 :- Fetching the movie having the maximum duration
df[df['duration']==f"{max_duration} min"]
```

```python
# 6. Find the content added in last 5 years 

df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y', errors='coerce')

# current date
today = pd.Timestamp.today()

# filter for last 5 years
df_recent = df[df['date_added'] >= (today - pd.DateOffset(years=5))]

df_recent
```

```python
# 7. Find all the movies and TV/Shows directed by Rajiv Chilaka 

df[df['director'].str.contains('rajiv chilaka',case=False , na=False)].count()
```

```python
# 8. List all the TV shows with more than 5 Seasons

# step 1 :- copy of the original datasets
df_tv=df.copy()

# step 2:- Fetching only TV shows and converting the duration object into Integer for comparison
df_tv.duration=df_tv[df_tv.type=='TV Show']['duration'].str.split(' ').str[0].astype('Int64')

#step 3:- fetching only unique shows having more than 5 seasons
greater=df_tv[df_tv.duration>5]['title'].unique()

df[df['title'].isin(greater)]
```

```python
# 9. Count the Number of content items in each Genre. 

df_new=df.copy()

# step 1:- converting the str into list of strings 
df_new['genres']=df_new['listed_in'].str.split(',')

#step 2 :- exploding the arrays into multiple duplicate rows and removing xtra spaces
df_new=df_new.explode('genres')
df_new['genres']=df_new['genres'].str.strip()

# step 3:- grouping it according to Genre and taking count of it
df_new.groupby('genres')['genres'].count().sort_values(ascending=False).head(7)
```

```python
# 10. Find each year and the average numbers of content release by India on Netflix  
# return top 5 year with Highest avg content release 
# select date_added from netflix 

# step 1:- Extracting rows where country belongs to India and keeping only years 
df_india=df[df.country.str.contains('india',case=False,na=False)]

df_india['year']=df_india['date_added'].dt.year

# step 2:- Grouping according to the year at which get added with frequency of it
year_num=df_india.groupby('year').size().reset_index(name='frequency')

total=year_num.frequency.sum()
year_num['avg_content']=round((year_num['frequency']/total)*100,2)

year_num.sort_values(by='avg_content',ascending=False).head(5)
```

```python
# 11. List all the movies that are documentaries 

df[df['listed_in'].str.contains('documentaries',case=False)]
```

```python
# 12. Find all the content without a director 

df[df.director.isnull()]
```

```python
# 13. Find how many movies actor 'Salman Khan' appeared in last 10 years

df.release_year=pd.to_datetime(df.release_year,format='%Y')
```

```python
df_salman=df[df.cast.str.contains('salman khan', case=False, na=False)]
today = pd.Timestamp.today()

df_salman[df_salman.release_year>(today-pd.DateOffset(years=10))]
```

```python
# 14. Find the top 10 actors who have appeared in the highest number of movies
# produced in India 

df_india=df[df.country.str.contains('india',case=False,na=False)]
df_india['actors']=df_india.cast.str.split(',')
df_new=df_india.explode('actors')
df_new['actors']=df_new.actors.str.strip()
df_new.groupby('actors')['show_id'].count().sort_values(ascending=False).head(10)
```

```python
# 15. Categorize the content based on the presence of the keyword like 'KILL' and 'VIOLENCE' 
#in the description field, label content as 'Bad' otherwise 'Good' . Count how many items falls in category.

df_rated=df.copy()

# df_rated['label'] = np.where(
#     df_rated['description'].str.contains('kill|violence', case=False, na=False),
#     'Bad',
#     'Good')


df_rated['label']=np.where(df_rated['description'].str.contains('kill|violence',case=False,na=False),'Bad','Good')

df_rated.groupby('label').size().reset_index(name='frequency')
```
